```python
import train_whisper

import torch
import pandas as pd
import time, json
model_size = 'base'

ver_start =1
start_epoch =13

mata_data = {
    'model_name' : f"openai/whisper-{model_size}",
    'path' : 'your path',
    'save_model_name' : f'models/whisper_{model_size}',
    'save_logging_file_name' : f'logging_{model_size}.json',
    'batch_size' : 4,
    'epochs' : 20,
    'start_epoch' : start_epoch, # 내가 추가함
    'ver_start' : ver_start, # 내가 추가함
    'lr' : 2e-5,
    'max_len' : 80,
    'device' : 'cuda'
}

# 불러오기
train_df = pd.read_csv(mata_data['path'] + 'pre_medical_audio_train.csv')
valid_df = pd.read_csv(mata_data['path'] + 'pre_medical_audio_valid.csv')
test_df = pd.read_csv(mata_data['path'] + 'pre_medical_audio_test.csv')

# train_df = train_df.iloc[:30]
# valid_df = valid_df.iloc[:30]

# 중간부터 시작할 경우

if ver_start == 0 and start_epoch == 0:
    res_dict = {
        'mata_data' : mata_data,
        'result_ver' : {}
    }
else:
    f = open(mata_data['path'] + mata_data['save_logging_file_name'], 'r')
    res_dict = json.load(f)
    res_dict['mata_data']['ver_start'] = ver_start
    res_dict['mata_data']['start_epoch'] = start_epoch
    f.close()

for ver in range(3):
    if ver < ver_start:
        continue
    print('--{}--'.format(ver))
    if str(ver) not in res_dict['result_ver'].keys():
        res_dict['result_ver'][str(ver)] = {
            'time' : None,
            # 'result' : None
            'result' : []
        }
    start = time.time()
    seed = ver
    res_dict = train_whisper.model_training(seed, mata_data, train_df, valid_df, test_df, res_dict)
    #res_dict['result_ver'][str(ver)]['result'] = res
    res_dict['result_ver'][str(ver)]['time'] = time.strftime('%X', time.localtime(time.time() - start))
    with open(mata_data['path'] + mata_data['save_logging_file_name'], 'w') as f:
        json.dump(res_dict, f, ensure_ascii=False, indent=4)
    f.close()

    start_epoch = 0
```
